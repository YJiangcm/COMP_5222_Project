```python
import torch
import os
import flair
from flair.data import Corpus
from flair.datasets import ColumnCorpus
from flair.embeddings import TokenEmbeddings, WordEmbeddings, StackedEmbeddings,CharacterEmbeddings, FlairEmbeddings
from typing import List
from flair.models import SequenceTagger

# Initialize Data

flair.device = torch.device("cuda:0")

columns = {0:'text'}

data_folder = '/content/drive/My Drive/5222_project/data/'

corpus: Corpus = ColumnCorpus(data_folder, columns,
                              train_file='fold_1234.txt',
                             dev_file='fold_5.txt',
                             test_file='fold_5.txt')
    
# 2. what tag do we want to predict?
tag_type = 'ner'

tag_dictionary = corpus.make_tag_dictionary(tag_type=tag_type)

# Load the model
from flair.models import SequenceTagger
model = SequenceTagger.load("/content/drive/My Drive/5222_project/output/best-model.pt")

# Write predictions in file

from flair.models import SequenceTagger
from flair.data import Sentence

#sentence = Sentence('Please extract information from the file')

# predict NER tags
#model.predict(sentence)

#print(sentence.to_tagged_string())

# print sentence with predicted tags

with open('bert-bio-fold_5.txt', 'w',encoding="utf-8") as f:
    for sentence in corpus.dev:
    
        model.predict(sentence)

        for token in sentence:
            f.write(f"{token.text}\t{token.get_tag('ner').value}" + "\n")
            
        f.write("\n")
```
